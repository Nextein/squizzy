```python
import dash
from dash import dcc, html
import plotly.express as px
import pandas as pd
from dash.dependencies import Input, Output

# Load and preprocess the data
file_path = './ilv_wallet1.csv'
data = pd.read_csv(file_path)

# Convert Quantity to numeric, removing commas
data['Quantity'] = data['Quantity'].str.replace(',', '').astype(float)

# Create the Dash app
app = dash.Dash(__name__)

app.layout = html.Div([
    html.H1("Wallet Transaction Dashboard"),
    
    # Total Transactions
    html.Div([
        html.H2("Total Transactions"),
        html.P(f"{data.shape[0]}")
    ], style={'textAlign': 'center'}),
    
    # Transaction Volume Over Time
    html.Div([
        dcc.Graph(id='volume-time', 
                  figure=px.line(data, x='DateTime (UTC)', y='Quantity', title='Transaction Volume Over Time'))
    ]),
    
    # Top Recipients and Senders
    html.Div([
        dcc.Graph(id='top-addresses',
                  figure=px.bar(data.melt(id_vars=['Transaction Hash', 'Quantity'], 
                                          value_vars=['From', 'To'], 
                                          var_name='Type', 
                                          value_name='Address'), 
                                x='Address', y='Quantity', color='Type', 
                                title='Top Recipients and Senders'))
    ]),
    
    # Transaction Types
    html.Div([
        dcc.Graph(id='transaction-types',
                  figure=px.pie(data, names='Method', title='Transaction Types'))
    ]),
    
    # Quantity Histogram
    html.Div([
        dcc.Graph(id='quantity-histogram',
                  figure=px.histogram(data, x='Quantity', title='Transaction Quantity Distribution'))
    ]),
])

if __name__ == '__main__':
    app.run_server(debug=True)
```

```python
import dash
from dash import dcc, html, Input, Output, State
import plotly.express as px
import pandas as pd
import dash_bootstrap_components as dbc

# Load and preprocess the data
file_path = './ilv_wallet1.csv'
data = pd.read_csv(file_path)

# Convert Quantity to numeric, removing commas
data['Quantity'] = data['Quantity'].str.replace(',', '').astype(float)

# Create the Dash app
app = dash.Dash(__name__, external_stylesheets=[dbc.themes.BOOTSTRAP])

app.layout = html.Div([
    html.H1("Wallet Transaction Dashboard"),
    
    # Total Transactions
    html.Div([
        html.H2("Total Transactions"),
        html.P(f"{data.shape[0]}")
    ], style={'textAlign': 'center'}),
    
    # Transaction Volume Over Time
    html.Div([
        dcc.Graph(id='volume-time', 
                  figure=px.line(data, x='DateTime (UTC)', y='Quantity', title='Transaction Volume Over Time'))
    ]),
    
    # Top Recipients and Senders
    html.Div([
        dcc.Graph(id='top-addresses',
                  figure=px.bar(data.melt(id_vars=['Transaction Hash', 'Quantity'], 
                                          value_vars=['From', 'To'], 
                                          var_name='Type', 
                                          value_name='Address'), 
                                x='Address', y='Quantity', color='Type', 
                                title='Top Recipients and Senders'))
    ]),
    
    # Transaction Types
    html.Div([
        dcc.Graph(id='transaction-types',
                  figure=px.pie(data, names='Method', title='Transaction Types'))
    ]),
    
    # Quantity Histogram
    html.Div([
        dcc.Graph(id='quantity-histogram',
                  figure=px.histogram(data, x='Quantity', title='Transaction Quantity Distribution'))
    ]),
    
    # Copy Wallet Addresses
    html.H2("Copy Wallet Addresses"),
    html.Div([
        html.Div([
            html.P(f"{address}", id=f"copy-{i}", style={'cursor': 'pointer', 'color': 'blue', 'text-decoration': 'underline'}),
            dcc.Clipboard(target_id=f"copy-{i}")
        ]) for i, address in enumerate(set(data['From'].tolist() + data['To'].tolist()))
    ], style={'display': 'flex', 'flexDirection': 'column', 'alignItems': 'flex-start'})
])

if __name__ == '__main__':
    app.run_server(debug=True)
```
